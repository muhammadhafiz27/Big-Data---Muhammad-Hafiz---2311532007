# **Import Library**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Load Dataset Titanic**

```python
df = sns.load_dataset('titanic')
df.head()
```

# **Eksplorasi Data**

Deskripsi Data

```python
df.shape
```

Statistik Deskriptif

```python
df.describe(include = "all")
```

Missing Value

```python
df.isnull().sum()
```

# **Visualisasi**

Distribusi Umur Penumpang

```python
sns.histplot(df['age'], kde=True)
plt.show()
```

Perbandingan Survival Berdasarkan Jenis Kelamin

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.show()
```

# **Latihan**

**Tampilkan 10 baris pertama dataset titanic**

```python
df.head(10)
```

**Hitung jumlah penumpang berdasarkan kelas**

```python
class_counts = df['class'].value_counts()
print("Jumlah penumpang berdasarkan kelas:")
print(class_counts)

sns.countplot(x='class', data=df)
plt.show()
```

**Visualisasikan perbandingan survival berdasarkan class**

```python
sns.countplot(x='class', hue='survived', data=df)
plt.show()
```

**Buat Kesimpulan singkat dari hasil eksplorasi awal**

Jadi bisa kita simpulkan dari visualisasi yang sudah kita buat bahwa jumlah penumpang terbanyak terdapat di Third class yaitu sebanyak 491 orang dan jumlah penumpang paling sedikit terdapat di Second class yaitu sebanyak 184 orang. Selain itu terdapat juga perbandingan jumlah penumpang yang selamat pada masing masing class dan didapatkan bahwa penumpang yang paling banyak tidak selamat terdapat di Third class dan penumpang yang paling banyak selamat di First class.
