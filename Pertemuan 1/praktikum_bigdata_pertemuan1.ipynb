**Deskripsikan secara singkat struktur dataset Titanic (fitur-fitur penting).**

Dataset Titanic berisi informasi tentang penumpang di kapal naas tersebut. Beberapa fitur penting meliputi:

*   **survived**: Menunjukkan apakah penumpang selamat (0 = Tidak, 1 = Ya).
*   **pclass**: Kelas penumpang (1, 2, atau 3).
*   **sex**: Jenis kelamin penumpang.
*   **age**: Usia penumpang.
*   **sibsp**: Jumlah saudara/pasangan di kapal.
*   **parch**: Jumlah orang tua/anak di kapal.
*   **fare**: Tarif yang dibayarkan oleh penumpang.
*   **embarked**: Pelabuhan keberangkatan (C = Cherbourg, Q = Queenstown, S = Southampton).
*   **class**: Sama seperti pclass, tetapi sebagai variabel kategorikal.
*   **who**: Laki-laki, perempuan, atau anak-anak.
*   **adult_male**: Boolean yang menunjukkan apakah penumpang adalah laki-laki dewasa.
*   **deck**: Dek tempat penumpang berada.
*   **embark_town**: Kota tempat penumpang naik kapal.
*   **alive**: Sama seperti survived, tetapi sebagai variabel kategorikal.
*   **alone**: Boolean yang menunjukkan apakah penumpang sendirian.

## Bagian 2 - Analisis Statistik Sederhana

**Hitung rata-rata umur (age) penumpang berdasarkan jenis kelamin.**

```python
average_age_by_sex = df.groupby('sex')['age'].mean()
print("Rata-rata umur penumpang berdasarkan jenis kelamin:")
print(average_age_by_sex)
```

**Hitung persentase penumpang yang selamat (survived) berdasarkan kelas (class).**

```python
survived_by_class = df.groupby('class')['survived'].value_counts(normalize=True).unstack() * 100
print("Persentase penumpang yang selamat berdasarkan kelas:")
print(survived_by_class)
```

**Cari jumlah penumpang yang tidak diketahui umurnya, lalu isi dengan rata-rata umur.**

```python
missing_age_count = df['age'].isnull().sum()
print(f"Jumlah penumpang dengan umur tidak diketahui: {missing_age_count}")

average_age = df['age'].mean()
df['age'].fillna(average_age, inplace=True)

print("\nJumlah missing values setelah diisi:")
display(df.isnull().sum())
```

## Bagian 3 - Visualisasi Data

**Buat histogram distribusi umur penumpang.**

```python
sns.histplot(df['age'], kde=True)
plt.title('Distribusi Umur Penumpang')
plt.xlabel('Umur')
plt.ylabel('Frekuensi')
plt.show()
```

**Buat barplot survival berdasarkan jenis kelamin.**

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah')
plt.show()
```

**Buat barplot survival berdasarkan kelas dan jenis kelamin (gabungan / grouped barplot).**

```python
g = sns.FacetGrid(df, col='sex', hue='survived')
g.map(sns.countplot, 'class', order=['First', 'Second', 'Third'])
g.add_legend()
plt.show()
```

**Buat visualisasi hubungan family_size (dari sibsp + parch + 1) dengan survival.**

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1
sns.countplot(x='family_size', hue='survived', data=df)
plt.title('Survival Berdasarkan Ukuran Keluarga')
plt.xlabel('Ukuran Keluarga')
plt.ylabel('Jumlah')
plt.show()
```
